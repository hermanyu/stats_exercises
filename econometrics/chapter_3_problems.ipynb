## Review the Concepts

#### 3.1) Explain the difference between the sample average $\overline{Y}$ and the population mean.

A: The sample average $\overline{Y}$ is the mean of a sample, i.e. a subset of the population. The population mean is the mean of the entire population.

<br>

#### 3.2) Explain the difference between an estimator and an estimate. Provide an example of each.

A: An estimator is a random variable, whose value depends on a given sample. An estimate is an actual value the estimator takes at a given sample. So for example, the sample average $\overline{Y}$ is an estimator, whose value will change as we draw different samples. For a given sample, we might have $\overline{Y}=100$ and that 100 is an estimate for the population mean.

<br>


#### 3.3) A population distribution has a mean of 10 and variance of 16. Determine the mean and variance of $\overline{Y}$ from an i.i.d. sample from this population for: (a) $n=10$ ; (b) $n=100$ ; (c) $ n=1000$. Relate your answer to the law of large numbers.

A: Since $\overline{Y}$ is an unbiased estimator for the population mean, we will always have $E(\overline{Y}) = 10$ no matter the same size $n$. For the variance, we have

$$Var(\overline{Y}) = \frac{Var(Y)}{n}$$

1. $Var(\overline{Y})=16/10 = 1.6$
2. $Var(\overline{Y})=16/100 = 0.16$
3. $Var(\overline{Y}) = 16/1000 = 0.016$

Notice that as $n\to \infty$, we have $Var(\overline{Y}) \to 0$, which says that the variability of the sample mean $\overline{Y}$ vanishes as the sample size gets larger. In other words, the sample mean $\overline{Y}$ converges in probability to the population, which is just the law of large numbers.

<br>


#### 3.4) What role does the Central Limit Theorem play in statistical hypothesis testing? In the construction of confidence intervales?

A: In order to conduct hypothesis tests or construct confidence intervals, we must be able to compute the CDF of the estimator in question. An estimator, by definition, is just a random variable and their behavior can be very "wild". However, in the case of the sample mean $\overline{Y}$, the Central Limit Theorem allows us to approximate $\overline{Y}$ as approximately normal, which allows to us to make computations using the CDF. This in turn allows us to compute p-values and construct confidence intervals.

<br>


#### 3.5) What is the difference between a null hypothesis and an alternative hypothesis? Among size, significance level, and power? Between a one-sided alternative hypothesis and a two-sided alternative hypothesis?

A: A null hypothesis is the default hypothesis which pre-supposes a parameter value (e.g. $\mu = 5$). The alternative hypothesis is a hypothesis which holds, when the null hypothesis fails. Alternative hypothesis come in 2 flavors: 2-sided alternative (e.g. $\mu \neq 5$) and 1-sided alternative (e.g. $\mu > 5$).

The **size** of a test is probability of making a Type II error: we reject the null hypothesis when it is true (i.e. the false positive rate). The **power** of the test is how likely we are to reject the null hypothesis when it is in-fact false (i.e. the true positive rate). The **significance level** $\alpha$ is the threshold with which we compare the p-value against and is used to decide where to accept or reject the null hypothesis. A higher $\alpha$ will mean rejection is more easily triggered, so we will reject the null hypothesis more often. Hence a higher $\alpha$ increases both size and power of the test. Conversely, a lower $\alpha$ means rejection is harder, so we will fail to reject the null hypothesis more often. Hence a lower $\alpha$ decreases both size and power of the test.

<br>

#### 3.6) Why does a confidence interval contain more information than the results of a single hypothesis test?

A: By definition, a confidence interval is the set of values $\mu_0$ for which the null hypothesis $\mu =\mu_0$ will fail to be rejected for a given sample at a given significance level. For example, suppose we have a sample with mean $\overline{Y}=10$. A **95% confidence interval** is the set of all values $\mu_0$ such that the following condition is true:

"If the null hypothesis is $\mu=\mu_0$, then we will fail to reject the null hypothesis $\mu=\mu_0$ at a 100%-95% = 5% = $\alpha$ significance level."

Heuristically, the 95% confidence interval is the set of all possible null hypothesis values $\mu_0$ that would not register as postive on a hypothesis test at 5% significance level. For this reason, a confidence interval can be thought of as containing the results of infinitely many hypothesis tests (one test for each value in the interval).

<br>

#### 3.7) Explain why the difference-of-means estimator, applied to data from a randomized experiment, is an estimator for the treatment effect.

A: Let us quickly remind ourselves of the estimators in question. The **casaul effect** or **treatment effect** is the parameter
$$E(Y|X=1) - E(Y|X=0)$$

In other words, the treatment effect is the difference in expected values of $Y$ between the treatment group $X=1$ and the control group $X=0$. 

Let $Y_0$ denote the random variable corresponding to the conditional distribution $Pr(Y|X=0)$. <br>
Let $Y_1$ denote the random variable corresponding to the conditional distribution $Pr(Y|X=1)$. 

Then the sample means $\overline{Y_0}$ and $\overline{Y_1}$ are unbiased estimators for $Y_0$ and $Y_1$ respectively. Therefore:
$$E(\overline{Y_0}) = E(Y_0)$$
$$E(\overline{Y_1}) = E(Y_1)$$

And therefore:
$$E(\overline{Y_1} - \overline{Y_0}) = E(\overline{Y_1})-E(\overline{Y_0})=E(Y_1)-E(Y_0) = E(Y|X=1) - E(Y|X=0)$$

Therefore the statistic $\overline{Y_1}-\overline{Y_0}$ is an unbiased estimator for the treatment effect $E(Y|X=1)-E(Y|X=0)$.

#### 3.1) In a population $\mu_Y=100$ and $\sigma^2_Y=43$. Use the CLT to answer the following questions: 

<ol>
     <li> In a sample of $n=100$, find $Pr(\overline{Y} <101)$. </li>
     <li> In a sample of $n=64$, find $Pr(101 < \overline{Y} <103)$. </li>
     <li> In a sample of $n=165$, find $Pr(\overline{Y}>98)$. </li>
</ol>


A: By the CLT, the sample mean $\overline{Y}$ is approximately normal with
$$\overline{Y} \sim N\left(100, \frac{43}{\sqrt{n}}\right)$$

```python
mu = 100
var = 43
```

```python
# 1. For n=100
n = 100

loc = mu
scale = np.sqrt(var/n)

X = stats.norm(loc=loc, scale=scale)

X.cdf(101)
```

```python
# 2. For n=64
n=64

loc = mu
scale = np.sqrt(var/n)

X = stats.norm(loc=loc, scale=scale)

X.cdf(103) - X.cdf(101)
```

```python
# 3. For n=165
n = 165

loc = mu
scale = np.sqrt(var/n)

X = stats.norm(loc=loc, scale=scale)

1 - X.cdf(98)
```

#### 3.2) Let $Y$ be a Bernoulli random variable with success probability $Pr(Y=1) = p$ and let $Y_1,\ldots,Y_n$ be i.i.d. draws from this distribution. Let $\hat{p}$ be the fraction of the successes (1's) in this sample.

<ol>
    <li> Show that $\hat{p} = \overline{Y}$. </li>
    <li> Show that $\hat{p}$ is an unbiased estimator of $p$. </li>
    <li> Show that $Var(\hat{p}) = p(1-p)/n$. </li>
</ol>


A: 

1) By definition, the estimator $\overline{Y}$ is the mean of the $Y_i$ and hence we have: $\overline{Y} = \frac{1}{n}\sum_{i=1}^n Y_i$. But notice that the value of $Y_i$ is either 1 or 0 and so the sum $\sum_{i=1}^n Y_i$ gives the total number of success. Therefore
$$\overline{Y} = \frac{1}{n}\sum_{i=1}^n Y_i = \frac{1}{n}\cdot (\text{  # of successess  } ) = \hat{p}$$

2) By the definition of a Bernoulli random variable, we have $E(Y) = p$ and by the previous part
$$E(\hat{p}) = E(\overline{Y}) = E(Y) = p$$

3) By the first part, we have $\hat{p}=\overline{Y}$. Also, since the $Y_i$ are i.i.d. the variance of the sum $\sum_{i=1}^n Y_i$ is additive and therefore:
$$ \begin{align*}
Var(\hat{p}) &= Var(\overline{Y}) \\
&= Var\left(\frac{1}{n}\sum_{i=1}^nY_i\right) \\
&=\frac{1}{n^2}\sum_{i=1}^nVar(Y_i) \\ 
&= \frac{1}{n^2}\sum_{i=1}^n Var(Y) \\
& = \frac{1}{n^2}\cdot n\cdot Var(Y) \\
& = \frac{1}{n}\cdot Var(Y)\\
& = \frac{1}{n} \cdot p(1-p)
\end{align*}$$

where we used the fact that the variance of a Bernoulli random variable $Y$ is precisely $p(1-p)$.

#### 3.3) In a survey of 400 likely voters, 215 responded that they would vote for the incumbent, and 185 responded that they would vote for the challenger. Let $p$ denote the fraction of all likely voters who preferred the incumbent at the time of the survey, and let $\hat{p}$  be the fraction of the survey respondents who preferred the incumbent.

<ol>
    <li> Use the survey results to estimate $p$. </li>
    <li> Use the estimator of the variance of $\hat{p}$, </li>
    <li> What is the $p$-value for the test $H_0: p=0.5$ vs $H_1:p\neq = 0.5$? </li>
    <li> What is the $p$-value for the test $H_0: p=0.5$ vs $H_1:p>0.5$? </li>
    <li> Why do the results from (3) and (4) differ? </li>
    <li> Did the survey contain statistically significant evidence that the incumbent was ahead of the challenger at the time of the survey? Explain </li>
</ol>

A: 

1) Using the sample proportion $hat{p}$ as an estimator for $p$, we have $p \approx \hat{p} = \frac{215}{400} = 0.5375$

```python
215/400
```

2. Using $\hat{p}(1-\hat{p})/n$ as an estimator for $Var(\hat{p})=p(1-p)/n$, we have:
$$SE(\hat{p}) = \sqrt{Var(\hat{p})} \approx \sqrt{\frac{\hat{p}(1-\hat{p})}{n}} = 0.0249295883439... $$

```python
p_hat = 215/400

np.sqrt(p_hat*(1-p_hat)/400)
```

3 ) By definition, the $p$-value of the sample statistic $215/400 = 0.5375$ is the probability of observing a result at least as extreme as our given sample. The phrase "at least as extreme" can have 2 meanings. "2-tailed" which means as extreme in either direction away from the mean. "1-tailed" which means as extreme in one specified direction away from the mean.

We have $\hat{p}=215/400 = 0.5375$. The (2-tailed) $p$-value for this sample estimate is the probability that $\hat{p}$ is at least as far away from the hypothesized mean $p=0.5$ as the observe sample statistic $0.5375$.
$$P(\hat{p} > 0.5375) + P(\hat{p} < 0.4625) $$

Let $Y$ be the random variable associated with randomly selecting a likely voter. Then $Y$ is a Bernoulli random variable with $Y=1$ meaning the selected voter supports the incumbent and $Y=0$ meaning otherwise. Let $Y_1,\ldots, Y_{400}$ be 400 i.i.d. draws of this Bernoulli random variable. By question 3.2 we have 
$$\hat{p} = \overline{Y}$$
In particular $\hat{p}$ is a sample mean, hence the Central Limit Theorem can be applied, and we can say $\hat{p}$ is approximately normally distributed:
$$\hat{p}\sim N\left(p \,\,, \sqrt{\frac{p(1-p)}{n}}\right)$$

Since we are hypothesizing that the true value of $p$ is $p=0.5$, we thus have:
$$\hat{p}\sim N(0.5, 0.025) $$

We can thus compute the desired probabilities $P(\hat{p} > 0.5375) + P(\hat{p}<0.4625)$

```python
p = 0.5
n = 400

se = np.sqrt(p*(1-p)/n)

X = stats.norm(loc=p, scale = se)
```

```python
prob1 = 1-X.cdf(0.5375)
prob2 = X.cdf(0.4625)

prob1, prob2, prob1+prob2
```

So the $p$-value for this test is $0.1336$

4 ) Again, the $p$-value of the sample statistic $215/400 = 0.5375$ is the probability of observing a result at least as extreme as our given sample. For 1-tailed hypothesis, "at least as extreme" means **at least as large** as our sample statistic of $0.5375$.

$$P(\hat{p}>0.5375)$$

5 ) The reason for the discrepancy comes down to the difference in the alternative hypotheses. The first hypothesis simply wishes to demonstrate that $p\neq 0.5$ and so sample statistics in both directions $\hat{p}<0.5$ and $\hat{p}>0.5$ would both be evidence for $p\neq 0.5$. Thus the phrase "at least as extreme" means any statistic as far away in both directions.

The second hypothesis wishes to demonstrate $p > 0.5$, so only sample statistics in the direction of $\hat{p}.0.5$ will provide evidence for this hypothesis. Thus the phrase "at least as extreme" means statistics which are as large as the sample value of $0.5375$.

More heuristically, we are using the sample statistic value $\hat{p}=p_0$ as evidence for the alternative hypothesis. It is MUCH easier to get samples with $\hat{p} \neq 0.5$ than it is to get samples with $\hat{p} > 0.5$. In fact the computations above show us that it is exactly TWICE as likely to get samples with $\hat{p} \neq 0.5$ than to get samples with $\hat{p} > 0.5$.
